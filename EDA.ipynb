## Join tables "anuario produccion agricola" 2019-2022

Cuadros 1-5

```python
import os
import pandas as pd
```

```python
files = os.listdir("./data_by_cuadro")
```

```python
# concat tables for diferent years
df = []
for name in set([f.split("_")[0] for f in files]):
      cuadros = []
      for file in [n for n in files if name in n ]:
      
            data = pd.read_csv("./data_by_cuadro/"+file)
            data.dropna(axis=1, how="all", inplace=True)
            # pivot: cultivos in one column
            data = data.melt(id_vars=["Región", "Año"], value_vars=[col for col in data.columns if col not in ["Región"]],
                  var_name="Cultivo", value_name=file.split("_")[0])
            cuadros.append(data)
      df.append(pd.concat(cuadros))
```

```python
# merge tables for region, anio and cultivo
print("joining tables ....")
data = df[0]
for i in df[1:]:
    data = pd.merge(data, i, on=["Región", "Año" ,"Cultivo"])

print("dimension result: ", data.shape)
```
