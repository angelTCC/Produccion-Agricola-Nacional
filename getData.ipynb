### Cuadro 3: Producción de principales cultivos, según región (t). 

```python
# load sheet -------------------------------
print("loading data ...")
data = pd.read_excel(io="../data/2730325-cuadros-en-excel-del-anuario-produccion-agricola-2022.xlsx", 
                     sheet_name="producción")

# recolled tables --------------------------------
print("searching tables ....")
list_tables = []
found_table = False
count = 0

for row in data.itertuples():

    if found_table == True:
        count +=1
        if row[1] != "Nacional": # no include nacional row
            df.loc[len(df)] = row[1:]
            if count == 27: # there is 27 row after the head
                list_tables.append(df)
                found_table = False
                count = 0      

    if row[1] == "Región":
        found_table = True
        df = pd.DataFrame(columns=row[1:])

print(len(list_tables), " tables found")

# join tables ---------------------------------------
print("joining tables ....")
df = list_tables[0]
for i in list_tables[1:]:
    df = pd.merge(df, i, on="Región")

print("dimension result: ", df.shape)

# remove nan columns  -------------------------------
print("removing nan columns")
nan_columns = []
for i in df.columns:
    if "nan" in i:
        nan_columns.append(i)
df.drop(columns=['nan_y', 'nan_x'], inplace=True)

# saveing -----------------------------
print("saving data ....")
df.to_csv("Produccion-principales-cultivos-segun-region.csv", index=False)
```
